```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots from the paper
iterations = np.array([5, 10, 15, 20, 25])
objective_values_A2_N2 = np.array([2, 4, 6, 8, 10])
objective_values_A3_N2 = np.array([2.5, 5, 7.5, 9, 10])
objective_values_A2_N4 = np.array([3, 5.5, 7.8, 9.5, 10])
objective_values_A3_N4 = np.array([3.5, 6, 8.5, 9.8, 10])

# Plot 1: Convergence of the proposed algorithm
plt.figure(figsize=(10, 6))
plt.plot(iterations, objective_values_A2_N2, marker='o', label='A=2, N=2')
plt.plot(iterations, objective_values_A3_N2, marker='s', label='A=3, N=2')
plt.plot(iterations, objective_values_A2_N4, marker='^', label='A=2, N=4')
plt.plot(iterations, objective_values_A3_N4, marker='d', label='A=3, N=4')
plt.xlabel('Number of Iterations')
plt.ylabel('Objective Value')
plt.title('Convergence of the Proposed Algorithm')
plt.legend()
plt.grid(True)
plt.show()

# Data for Secrecy Rate vs. Target SNR
snr_targets = np.array([-5, 0, 5, 10, 15])
secrecy_rate_MA2 = np.array([0.1, 0.5, 1.0, 1.5, 2.0])
secrecy_rate_FPA2 = np.array([0.05, 0.4, 0.8, 1.2, 1.6])
secrecy_rate_MA3 = np.array([0.2, 0.7, 1.3, 1.8, 2.3])
secrecy_rate_FPA3 = np.array([0.1, 0.5, 1.0, 1.4, 1.9])
secrecy_rate_MA4 = np.array([0.3, 0.9, 1.5, 2.0, 2.5])
secrecy_rate_FPA4 = np.array([0.15, 0.6, 1.1, 1.6, 2.0])

# Plot 2: Secrecy Rate vs. Target SNR
plt.figure(figsize=(10, 6))
plt.plot(snr_targets, secrecy_rate_MA2, marker='o', label='MA, N=2')
plt.plot(snr_targets, secrecy_rate_FPA2, marker='s', label='FPA, N=2')
plt.plot(snr_targets, secrecy_rate_MA3, marker='^', label='MA, N=3')
plt.plot(snr_targets, secrecy_rate_FPA3, marker='d', label='FPA, N=3')
plt.plot(snr_targets, secrecy_rate_MA4, marker='x', label='MA, N=4')
plt.plot(snr_targets, secrecy_rate_FPA4, marker='*', label='FPA, N=4')
plt.xlabel('Target SNR [dB]')
plt.ylabel('Secrecy Rate [bps/Hz]')
plt.title('Secrecy Rate vs. Target SNR')
plt.legend()
plt.grid(True)
plt.show()

# Data for Secrecy Rate vs. Region Size
region_sizes = np.array([1, 2, 3, 4, 5, 6])
secrecy_rate_MA = np.array([0.8, 1.0, 1.4, 1.8, 2.0, 2.2])
secrecy_rate_FPA = np.array([0.5, 0.7, 1.0, 1.2, 1.3, 1.4])

# Plot 3: Secrecy Rate vs. Region Size
plt.figure(figsize=(10, 6))
plt.plot(region_sizes, secrecy_rate_MA, marker='o', label='MA')
plt.plot(region_sizes, secrecy_rate_FPA, marker='s', label='FPA')
plt.xlabel('Normalized Region Size A/λ')
plt.ylabel('Secrecy Rate [bps/Hz]')
plt.title('Secrecy Rate vs. Region Size')
plt.legend()
plt.grid(True)
plt.show()

```
